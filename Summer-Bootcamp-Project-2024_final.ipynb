# 10-	Do the necessary data cleaning steps like dropping duplicates, unnecessary columns, null value imputation, outliers treatment etc.

```python
#check for outliers
for i in data.columns:
    if data[i].dtype != 'object':
        sns.boxplot(data=data,x=i)
        plt.show()
```

```python
median1=data["Electric Range"].median()
mode1=data["County"].mode().values[0]
mode2=data["City"].mode().values[0]
mode3=data["Postal Code"].mode().values[0]
mode4=data["Make"].mode().values[0]
mode5=data["Model"].mode().values[0]
mode6=data["Electric Vehicle Type"].mode().values[0]
mode7=data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].mode().values[0]
mode8=data["Legislative District"].mode().values[0]
mode9=data["Vehicle Location"].mode().values[0]
mode10=data["Electric Utility"].mode().values[0]

data["Electric Range"].replace(np.nan,median1,inplace=True)
data["County"]=data["County"].replace(np.nan,mode1)
data["City"]=data["City"].replace(np.nan,mode2)
data["Postal Code"]=data["Postal Code"].replace(np.nan,mode3)
data["Make"]=data["Make"].replace(np.nan,mode4)
data["Model"]=data["Model"].replace(np.nan,mode5)
data["Electric Vehicle Type"]=data["Electric Vehicle Type"].replace(np.nan,mode6)
data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"]=data["Clean Alternative Fuel Vehicle (CAFV) Eligibility"].replace(np.nan,mode7)
data["Legislative District"]=data["Legislative District"].replace(np.nan,mode8)
data["Vehicle Location"]=data["Vehicle Location"].replace(np.nan,mode9)
data["Electric Utility"]=data["Electric Utility"].replace(np.nan,mode10)
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range, upper_range
```

```python
for i in data.columns:
    if data[i].dtype !='object':
        lr,ur=remove_outlier(data[i])
        data[i]=np.where(data[i]>ur,ur,data[i])
        data[i]=np.where(data[i]<lr,lr,data[i])
```

# 17.	Vehicle Location Analysis:
•	Analyze the geographic coordinates to determine any clusters of electric vehicles in certain areas of Washington state.


```python
if 'Vehicle Location' in df.columns:
    # Drop rows with NaN in 'Vehicle Location'
    df = df.dropna(subset=['Vehicle Location'])
    
    # Extract 'Longitude' and 'Latitude' from 'Vehicle Location'
    def extract_coordinates(point):
        try:
            point = point.replace('POINT (', '').replace(')', '')
            lon, lat = point.split()
            return pd.Series([float(lat), float(lon)])
        except Exception as e:
            return pd.Series([None, None])
    
    df[['Latitude', 'Longitude']] = df['Vehicle Location'].apply(extract_coordinates)
    
    # Drop rows with NaN values in 'Latitude' or 'Longitude'
    df = df.dropna(subset=['Latitude', 'Longitude'])
    
    # Plot the data to visualize clusters
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x='Longitude', y='Latitude', data=df)
    plt.title('Geographic Distribution of Electric Vehicles')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()
else:
    print("The 'Vehicle Location' column is not present in the DataFrame.")
```

#### Figure 10: Geographic Distribution of Electric Vehicles

# Observation:-
    -The plot shows a significant cluster of electric vehicles around the 47-50°N latitude and 120-125°W longitude region, suggesting a concentration in Washington state.
